## **Clear ML**

**ClearML** представляет собой платформу для отслеживания, управления и улучшения процесса машинного обучения. Позволяет исследователям эффективно управлять экспериментами, отслеживать модели, а также собирать и анализировать данные обучения.

Основные возможности ClearML включают в себя:

1. **Отслеживание экспериментов:** Платформа ClearML предоставляет возможность систематического отслеживания экспериментов, включая параметры, метрики, логи и другие важные данные.

2. **Управление моделями:** ClearML помогает управлять моделями на всех этапах их жизненного цикла — от разработки и обучения до внедрения в продакшн. Это включает в себя версионирование моделей, их сравнение и выбор наилучшей модели.

3. **Визуализация и анализ данных:** Платформа предоставляет инструменты для визуализации результатов обучения и анализа данных.

## **YOLOv8**

**YOLOv8** - это последняя версия модели YOLO (You Only Look Once), разработанная компанией Ultralytics. Модель YOLOv8 разработана для быстрого, точного и простого использования, что делает ее отличным выбором для широкого круга задач по обнаружению объектов, сегментации и классификации изображений.

1. **Высокая скорость:** YOLOv8 обеспечивает высокую скорость работы даже на более высоких разрешениях изображений.
2. **Эффективность и точность:** YOLOv8 стремится достичь баланса между скоростью работы и точностью. Она предоставляет достаточно высокую точность для многих приложений компьютерного зрения.
3. **Простота использования:** Одним из принципов YOLO является его простота. YOLOv8 продолжает это традицию, обеспечивая относительно простой интерфейс для обучения и использования модели.

YOLOv8 включает в себя различные дополнительные функции и оптимизации, такие как автоматическое масштабирование для разных разрешений изображений, поддержка различных бэкбонов (например, CSPDarknet53) и другие.

```python
from clearml import Dataset

# Создаем новый датасет в ClearML
dataset = Dataset.create(
    dataset_project='Skin',  # Имя проекта в ClearML
    dataset_name='Skin_Rak_1'  # Имя вашего датасета
)

# Добавляем файлы в датасет
dataset.add_files('/content/datasets/rak_skin-1')

# Загружаем датасет в хранилище ClearML
dataset.upload()

# Завершаем создание датасета
dataset.finalize()
```
